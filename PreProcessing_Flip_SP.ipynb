# Flip Based Pre-processing for Fairness (SP)

|Disparity $d$|S.Attr. $s_i$|Label. $y_{i}$|Flip $y_i$?|
|-------------|-------------|--------------|-----------|
|d>0          |$s_i$=0      |$y_i$<0.5     |N          |
|d>0          |$s_i$=0      |$y_i$>0.5     |Y          |
|d>0          |$s_i$=1      |$y_i$<0.5     |Y          |
|d>0          |$s_i$=1      |$y_i$>0.5     |N          |
|d<0          |$s_i$=0      |$y_i$<0.5     |Y          |
|d<0          |$s_i$=0      |$y_i$>0.5     |N          |
|d<0          |$s_i$=1      |$y_i$<0.5     |N          |
|d<0          |$s_i$=1      |$y_i$>0.5     |Y          |

$\textbf{FLAG}=d (s_i-0.5) (\hat{y}_i-0.5) < 0$

```python
from FaroPreFlip import flip, draw
```

```python
res=flip('adult','race',n_epoch=1000,ascending=True)
```

```python
draw(res)
```
